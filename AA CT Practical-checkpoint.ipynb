```python
ac2 = AgglomerativeClustering(n_clusters = 2)
ac2
```

```python
ac2.fit_predict(X_scaled)
```

```python
silhouette_score(X_scaled, ac2.labels_)
```

```python
k_range = range(2,10)
silhouette_scores =[]

for i in k_range:
    ac_i = AgglomerativeClustering(n_clusters = i,linkage='ward')
    silhouette_scores.append(silhouette_score(X_scaled, ac_i.fit_predict(X_scaled)))
```

```python
silhouette_scores
```

```python
plt.bar(k_range, silhouette_scores) 
plt.xlabel('Number of clusters', fontsize = 20) 
plt.ylabel('Silouette Score', fontsize = 20)
plt.axis([1, 11, 0.1, 0.4])
plt.show() 
```

```python
sse = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, n_init=20, random_state=1)
    kmeans.fit(X_scaled)
    sse.append(kmeans.inertia_)
plt.plot(range(1, 11), sse, 'b-*')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.show()
```
